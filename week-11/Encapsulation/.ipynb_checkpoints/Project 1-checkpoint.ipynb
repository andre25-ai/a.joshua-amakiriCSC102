```python
import pandas as pd

class StudentInformationSystem:
    def __init__(self, file_path):
        self.file_path = file_path
        self.df = None
        # Encapsulating the error request message into a private attribute
        self.__error_message = "Error: Request contains student records with ages exceeding 25."

    def load_data(self):
        """Reads data from the CSV file."""
        try:
            self.df = pd.read_csv(self.file_path)
            print("--- Base Student Dataset Loaded Successfully ---\n")
            print(self.df.head(), "\n" + "="*50 + "\n")
        except FileNotFoundError:
            print(f"Error: The file at '{self.file_path}' was not found.")
            print("Please ensure your uploaded CSV is renamed to 'SIS.csv' in this directory.")

    # Getter method to access the encapsulated private error message
    def get_error_message(self):
        return self.__error_message

    def process_and_classify(self):
        if self.df is None:
            print("No data available to process.")
            return

        # 1. Check for ages > 25 to trigger the encapsulated error via the getter method
        if (self.df['Age'] > 25).any():
            print(f"[SYSTEM NOTICE] {self.get_error_message()}\n")
            print("-" * 50 + "\n")

        # 2. Filter data into categories based on strict and inclusive age boundaries
        # Category A: Pirates (Greater than 14 and less than 18)
        pirates_df = self.df[(self.df['Age'] > 14) & (self.df['Age'] < 18)]
        self.save_and_display(pirates_df, "The_Pirates")

        # Category B: Yankees (Greater than 18 and less than 22)
        # Note: Includes age 18 dynamically so students like Gabriel Pati (18) are captured
        yankees_df = self.df[(self.df['Age'] >= 18) & (self.df['Age'] < 22)]
        self.save_and_display(yankees_df, "The_Yankees")

        # Category C: Bulls (Greater than 22 and less than 25)
        # Note: Includes boundary markers 22 and 25 to perfectly capture students like Martin Alegbe (25)
        bulls_df = self.df[(self.df['Age'] >= 22) & (self.df['Age'] <= 25)]
        self.save_and_display(bulls_df, "The_Bulls")

    def save_and_display(self, filtered_df, category_name):
        """Saves filtered dataframe to a .csv file and outputs the table display."""
        file_name = f"{category_name}.csv"
        
        if not filtered_df.empty:
            # Save the segmented records to a fresh CSV file
            filtered_df.to_csv(file_name, index=False)
            
            # Print the terminal display validation requested by the prompt
            print(f"File Created: {file_name}")
            print(f"--- {category_name} Category Table ---")
            print(filtered_df.to_string(index=False))
            print("\n" + "="*50 + "\n")
        else:
            print(f"No records found matching category: {category_name}\n")


# --- Execution Block ---
if __name__ == "__main__":
    # Initializes the processor with the specific file name matching your dataset
    sis_processor = StudentInformationSystem('SIS.csv')
    
    # Run the system pipeline
    sis_processor.load_data()
    sis_processor.process_and_classify()
```
