```python
import tkinter as tk
from tkinter import ttk
from abc import ABC, abstractmethod

# Abstract Base Class
class External_Vendors(ABC):
    @abstractmethod
    def menu(self):
        pass

# Subclasses representing each vendor and their unique menu data
class Cooperative_Cafeteria(External_Vendors):
    def menu(self):
        return [
            ("Jollof Rice and Stew", "200"),
            ("White Rice and Stew", "200"),
            ("Fried Rice", "200"),
            ("Salad", "100"),
            ("Plantain", "100")
        ]

class Faith_Hostel_Cafeteria(External_Vendors):
    def menu(self):
        return [
            ("Fried Rice", "400"),
            ("White Rice and Stew", "400"),
            ("Jollof Rice", "400"),
            ("Beans", "200"),
            ("Chicken", "1000")
        ]

class Student_Centre_Cafeteria(External_Vendors):
    def menu(self):
        return [
            ("Chicken Fried Rice", "800"),
            ("Pomo Sauce", "300"),
            ("Spaghetti Jollof", "500"),
            ("Amala/Ewedu", "500"),
            ("Semo with Eforiro Soup", "500")
        ]

# GUI Management Class
class VendorApp:
    def __init__(self, root):
        self.root = root
        self.root.title("PAU External Food Vendors Portal")
        self.root.geometry("550x450")
        self.root.configure(bg="#1A1A18")  # Premium dark background
        
        # Mapping dropdown options to Class instances
        self.vendor_mapping = {
            "Cooperative Cafeteria": Cooperative_Cafeteria(),
            "Faith Hostel Cafeteria": Faith_Hostel_Cafeteria(),
            "Student Centre Cafeteria": Student_Centre_Cafeteria()
        }
        
        self.create_widgets()
        
    def create_widgets(self):
        # Title Label
        title_lbl = tk.Label(
            self.root, 
            text="PAU Food Vendor Portal", 
            font=("Helvetica", 16, "bold"), 
            bg="#1A1A18", 
            fg="#D4A843" # Gold accent
        )
        title_lbl.pack(pady=15)
        
        # Dropdown selection frame
        select_frame = tk.Frame(self.root, bg="#1A1A18")
        select_frame.pack(pady=10)
        
        dropdown_lbl = tk.Label(
            select_frame, 
            text="Select Vendor: ", 
            font=("Helvetica", 11), 
            bg="#1A1A18", 
            fg="#FFFFFF"
        )
        dropdown_lbl.pack(side=tk.LEFT, padx=5)
        
        # Combobox setup
        self.vendor_var = tk.StringVar()
        self.dropdown = ttk.Combobox(
            select_frame, 
            textvariable=self.vendor_var, 
            values=list(self.vendor_mapping.keys()), 
            state="readonly", 
            width=25
        )
        self.dropdown.pack(side=tk.LEFT, padx=5)
        self.dropdown.bind("<<ComboboxSelected>>", self.display_menu)
        
        # Treeview frame for displaying menu items cleanly
        self.table_frame = tk.Frame(self.root, bg="#1A1A18")
        self.table_frame.pack(pady=20, fill=tk.BOTH, expand=True, padx=30)
        
        # Configure a modern look for Treeview elements
        style = ttk.Style()
        style.theme_use("clam")
        style.configure("Treeview.Heading", font=("Helvetica", 11, "bold"), background="#D4A843", foreground="#1A1A18")
        style.configure("Treeview", font=("Helvetica", 10), rowheight=25)
        
        # Creating Treeview columns
        self.tree = ttk.Treeview(self.table_frame, columns=("Main Meal", "Price"), show="headings")
        self.tree.heading("Main Meal", text="Main Meal")
        self.tree.heading("Price", text="Price (₦)")
        self.tree.column("Main Meal", width=300, anchor=tk.W)
        self.tree.column("Price", width=120, anchor=tk.CENTER)
        self.tree.pack(fill=tk.BOTH, expand=True)

    def display_menu(self, event=None):
        # Clear existing table data
        for row in self.tree.get_children():
            self.tree.delete(row)
            
        selected_vendor_name = self.vendor_var.get()
        vendor_instance = self.vendor_mapping.get(selected_vendor_name)
        
        if vendor_instance:
            # Polymorphic method execution retrieving the unique item lists
            menu_items = vendor_instance.menu()
            for item, price in menu_items:
                self.tree.insert("", tk.END, values=(item, f"₦ {price}"))

if __name__ == "__main__":
    root = tk.Tk()
    app = VendorApp(root)
    root.mainloop()
```
