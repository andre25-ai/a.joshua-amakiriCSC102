```python
import tkinter as tk
from tkinter import messagebox
import random

class EmployeeSystem:
    def __init__(self, root):
        self.root = root
        self.root.title("Mrs. Jane's Logistics System")
        self.root.geometry("400x300")
        
        # Data
        self.employees = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", "Andrew Ugwu", 
            "Stella Mankinde", "Jane Akibo", "Ago James", "Michell Taiwo", "Abraham Jones", 
            "Nicole Anide", "Kosi Korso", "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
        ]
        self.tasks = ["Loading", "Transporting", "Reviewing Orders", "Customer Service", "Delivering Items"]
        self.attendance_log = {}

        # GUI Elements
        self.label = tk.Label(root, text="Enter Employee Name:", font=("Arial", 12))
        self.label.pack(pady=10)
        
        self.name_entry = tk.Entry(root, font=("Arial", 12), width=25)
        self.name_entry.pack(pady=5)
        
        self.submit_btn = tk.Button(root, text="Verify & Check-In", command=self.check_employee, bg="green", fg="white")
        self.submit_btn.pack(pady=15)

    def check_employee(self):
        entered_name = self.name_entry.get().strip()
        
        # Case-insensitive or direct match validation
        matching_employee = next((emp for emp in self.employees if emp.lower() == entered_name.lower()), None)
        
        if matching_employee:
            self.take_attendance(matching_employee)
            self.assign_task(matching_employee)
        else:
            self.refuse_access()

    def take_attendance(self, employee_name):
        self.attendance_log[employee_name] = "Present"
        print(f"Attendance Logged: {employee_name} is Present.")

    def assign_task(self, employee_name):
        assigned_task = random.choice(self.tasks)
        messagebox.showinfo("Access Granted", f"Welcome, {employee_name}!\n\nYour task today: {assigned_task}")

    def refuse_access(self):
        messagebox.showerror("Access Denied", "Access Denied: Name not found in the employee database.")

# Run the App
if __name__ == "__main__":
    root = tk.Tk()
    app = EmployeeSystem(root)
    root.mainloop()
```
