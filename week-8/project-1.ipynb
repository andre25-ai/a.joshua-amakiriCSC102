```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox

# --- OOP Backend Classes ---
class Zenith:
    def __init__(self, employee_name):
        self.employee_name = employee_name

    def mutual_services(self):
        return ["• Lines of credit", "• Investment management and accounts", "• Insurance"]

    def unique_services(self):
        return []

class RetailBanking(Zenith):
    def unique_services(self):
        return ["• Retirement and education accounts", "• Loans and mortgages", "• Checking and saving"]

class GlobalBanking(Zenith):
    def mutual_services(self):
        return [] # Global shares none of the retail/commercial mutual services listed
    def unique_services(self):
        return [
            "• Multi-currency management services and products",
            "• Foreign currency accounts",
            "• Foreign currency credit cards",
            "• Transborder advisory services",
            "• Liquidity management"
        ]

class CommercialBanking(Zenith):
    def unique_services(self):
        return ["• Advisory services"]


# --- UI Frontend Class ---
class ZenithBankApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Zenith Bank Nigeria - Division Portal")
        self.root.geometry("500x550")
        self.root.config(bg="#f4f4f4")

        # Title Banner
        title_label = tk.Label(root, text="ZENITH BANK SERVICE PORTAL", font=("Helvetica", 14, "bold"), bg="#1e3d59", fg="white", pady=10)
        title_label.pack(fill="x")

        # Main Container Frame
        main_frame = tk.Frame(root, bg="#f4f4f4", padx=20, pady=15)
        main_frame.pack(fill="both", expand=True)

        # Employee Name Inputs
        tk.Label(main_frame, text="Employee Name:", font=("Arial", 11, "bold"), bg="#f4f4f4").pack(anchor="w", pady=2)
        self.name_entry = tk.Entry(main_frame, font=("Arial", 11), width=40)
        self.name_entry.pack(fill="x", pady=5)
        self.name_entry.bind("<KeyRelease>", self.auto_detect_division) # Auto-detects division as they type!

        # Division Dropdown Selection
        tk.Label(main_frame, text="Select Division:", font=("Arial", 11, "bold"), bg="#f4f4f4").pack(anchor="w", pady=2)
        self.divisions = ["Retail Banking", "Global Banking", "Commercial Banking"]
        self.division_dropdown = ttk.Combobox(main_frame, values=self.divisions, state="readonly", font=("Arial", 11))
        self.division_dropdown.pack(fill="x", pady=5)

        # Action Button
        self.fetch_btn = tk.Button(main_frame, text="Fetch Services Rendered", command=self.display_services, 
                                   bg="#ff6e40", fg="white", font=("Arial", 11, "bold"), height=2, cursor="hand2")
        self.fetch_btn.pack(fill="x", pady=15)

        # Output Text Area
        tk.Label(main_frame, text="Rendered Division Services:", font=("Arial", 11, "bold"), bg="#f4f4f4").pack(anchor="w", pady=2)
        self.output_display = tk.Text(main_frame, font=("Courier New", 10), height=12, bg="white", relief="solid", bd=1, padx=10, pady=10)
        self.output_display.pack(fill="both", expand=True)
        self.output_display.config(state="disabled")

    def auto_detect_division(self, event):
        """Automatically selects the division if default employee names are typed."""
        name = self.name_entry.get().strip().lower()
        if name == "mary":
            self.division_dropdown.set("Retail Banking")
        elif name == "agatha":
            self.division_dropdown.set("Global Banking")
        elif name == "noel":
            self.division_dropdown.set("Commercial Banking")

    def display_services(self):
        name = self.name_entry.get().strip()
        division = self.division_dropdown.get()

        if not name or not division:
            messagebox.showwarning("Incomplete Data", "Please verify both Employee Name and Division fields.")
            return

        # Instantiate the correct class mapping using Polymorphism
        if division == "Retail Banking":
            emp_object = RetailBanking(name)
        elif division == "Global Banking":
            emp_object = GlobalBanking(name)
        else:
            emp_object = CommercialBanking(name)

        # Extract data through parent inheritance and child polymorphism
        mutual = emp_object.mutual_services()
        unique = emp_object.unique_services()

        # Update Text Interface
        self.output_display.config(state="normal")
        self.output_display.delete("1.0", tk.END)
        
        greeting = f"User: {emp_object.employee_name}\nDivision: {division}\n"
        self.output_display.insert(tk.END, greeting + "="*40 + "\n\n")

        if mutual:
            self.output_display.insert(tk.END, "Mutual Services Provided:\n")
            for service in mutual:
                self.output_display.insert(tk.END, f"{service}\n")
            self.output_display.insert(tk.END, "\n")

        if unique:
            self.output_display.insert(tk.END, "Exclusive/Unique Services Provided:\n")
            for service in unique:
                self.output_display.insert(tk.END, f"{service}\n")
                
        self.output_display.config(state="disabled")


if __name__ == "__main__":
    root = tk.Tk()
    app = ZenithBankApp(root)
    root.mainloop()
```
