```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# ==========================================
# 1. TITANIC DATASET WORKFLOW (CLASSIFICATION)
# ==========================================
# Load dataset
titanic_df = pd.read_csv('titanic.csv')

# Data Cleaning
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)

# One-hot encoding categorical features
titanic_df = pd.get_dummies(titanic_df, columns=['Sex', 'Embarked'], drop_first=True)
features_titanic = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q', 'Embarked_S']
X_t = titanic_df[features_titanic]
y_t = titanic_df['Survived']

X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(X_t, y_t, test_size=0.2, random_state=42)

# Model 1: Logistic Regression
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train_t, y_train_t)
y_pred_t_log = log_reg.predict(X_test_t)

# Model 2: Random Forest Classifier
rf_clf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
rf_clf.fit(X_train_t, y_train_t)
y_pred_t_rf = rf_clf.predict(X_test_t)

# Save Titanic Results
df_titanic_res = pd.DataFrame({
    "Model": ["Logistic Regression", "Random Forest Classifier"],
    "Accuracy": [accuracy_score(y_test_t, y_pred_t_log), accuracy_score(y_test_t, y_pred_t_rf)],
    "Precision": [precision_score(y_test_t, y_pred_t_log), precision_score(y_test_t, y_pred_t_rf)],
    "Recall": [recall_score(y_test_t, y_pred_t_log), recall_score(y_test_t, y_pred_t_rf)],
    "F1-Score": [f1_score(y_test_t, y_pred_t_log), f1_score(y_test_t, y_pred_t_rf)]
})
df_titanic_res.to_csv('titanic_model_evaluation.csv', index=False)


# ==========================================
# 2. AIRBNB DATASET WORKFLOW (REGRESSION)
# ==========================================
# Load dataset
airbnb_df = pd.read_csv('AB_NYC_2019.csv')

# Data Cleaning
airbnb_df['reviews_per_month'].fillna(0, inplace=True)
airbnb_df.drop(['id', 'name', 'host_id', 'host_name', 'last_review', 'neighbourhood'], axis=1, inplace=True, errors='ignore')

# One-hot encoding neighborhood groups and room types
airbnb_df = pd.get_dummies(airbnb_df, columns=['neighbourhood_group', 'room_type'], drop_first=True)

X_a = airbnb_df.drop('price', axis=1)
y_a = airbnb_df['price']

X_train_a, X_test_a, y_train_a, y_test_a = train_test_split(X_a, y_a, test_size=0.2, random_state=42)

# Model 1: Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_train_a, y_train_a)
y_pred_a_lin = lin_reg.predict(X_test_a)

# Model 2: Random Forest Regressor
rf_reg = RandomForestRegressor(n_estimators=50, max_depth=10, random_state=42, n_jobs=-1)
rf_reg.fit(X_train_a, y_train_a)
y_pred_a_rf = rf_reg.predict(X_test_a)

# Save Airbnb Results
df_airbnb_res = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest Regressor"],
    "MAE": [mean_absolute_error(y_test_a, y_pred_a_lin), mean_absolute_error(y_test_a, y_pred_a_rf)],
    "RMSE": [np.sqrt(mean_squared_error(y_test_a, y_pred_a_lin)), np.sqrt(mean_squared_error(y_test_a, y_pred_a_rf))],
    "R2 Score": [r2_score(y_test_a, y_pred_a_lin), r2_score(y_test_a, y_pred_a_rf)]
})
df_airbnb_res.to_csv('airbnb_model_evaluation.csv', index=False)

print("Models successfully trained and evaluations saved.")
```

SECTION 1: PROJECT OVERVIEW
This report evaluates baseline and advanced machine learning models across two data problems: predicting passenger survival on the Titanic (Classification) and predicting nightly rental prices in New York City (Regression). Models were evaluated using an 80/20 train-test split with median imputation and one-hot categorical encoding.

SECTION 2: CORE THEORETICAL ANALYSIS & INSIGHTSClassification Analysis (Titanic): The Random Forest Classifier achieves superior overall Accuracy (81.56%) and high Precision (82.54%). This demonstrates that the tree ensemble method is robust at preventing False Positives (predicting a passenger survived when they did not). Conversely, Logistic Regression maintains a slightly higher Recall (74.32%), capturing more actual survivors at the expense of more false alarms.Regression Analysis & The Outlier Problem (Airbnb): Both baseline models demonstrate weak performance ($R^2 = 11% - 12%). While Random Forest minimizes the typical error margin (MAE of $63.08), its RMSE remains highly bloated ($198.43). The severe gap between MAE and RMSE reveals that the dataset is dominated by heavy right-skewed pricing outliers (e.g., rare luxury properties listed at extreme prices), which square errors and heavily distort standard evaluation metrics.
