PSEUDOCODE
1. BEGIN
2. SCAN ITEM PRICES USING A BARCODE READER
3. CALCULATE TOTAL PRICE OF INPUTS
4. SELECT PAYMENT METHOD (CASH/CARD)
5. IF CASH:
   GIVE CASH AND COLLECT CHANGE
   IF CARD:
   SWIPE CARD
6. IS PAYMENT SUCCESSFUL ? (YES/NO)
   IF NO:
   ERROR MESSAGE THEN GO BACK TO PAYMENT METHOD
   IF YES:
   PROCESS COMPLETE
   COLLECT RECEIPT
7. END

ALGORTIHM
1. START
2. INPUT NAME1
3. INPUT NAME2
4. INPUT AGE1
5. INPUT AGE2
6. CREATE A TEMPORARY VARIABLE FOR AGE1
7. THEN EQUATE AGE1 TO AGE2
8. EQUATE AGE2 TO THE TEMPORARY VARIABLE
9. PRINT NAME1 AND AGE1
10. PRINT NAME2 AND AGE2
11. END
