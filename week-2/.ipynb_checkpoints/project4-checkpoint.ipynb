```python
print("\n Annual Tax Revenue Calculator for Izfin technolgy")

exp = input("Do you have years of experience ? (yes/no): ")

if exp.lower() == "yes":
            years = float(input("How many years of experience"))
            age = int(input("How old are you: "))

            if years > 25 and age >= 55:
                      atr = 5600000
            elif years > 20 and age >= 45:
                      atr = 4480000
            elif years > 10 and age >= 35:
                      atr = 1500000
            else:
                      atr = 550000
            print("The Annual Tax Revenue (ATR) for this staff is: N ",atr)
else:
   print("Experience is needed to be eligible for Annual Tax Revenue (ATR)")
                      
            


                
          
```
