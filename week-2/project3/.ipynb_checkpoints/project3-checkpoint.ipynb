```python
import pandas as pd

print("Program to represent the given student data in a tabular form")
#girls data
g_names = ( "Evelyn", "Jessica", "Somto", "Edith", "Liza", "Madonna", "Waje", "Tola", "Aisha", "Latifa")
g_age = (17, 16, 17, 18, 16, 18, 17, 20, 19, 17)
g_height = (5.5, 6.0 , 5.4, 5.9, 5.6, 5.5, 6.1, 6.0, 5.7, 5.5) 
g_scores = (80, 85, 70, 60, 76, 66, 87, 95, 50, 49)
g_gender = ["Female"] * 10
#boys data
b_names = ("Chinedu", "Liam", "Wale", "Gbenga", "Abiola", "Kola", "Kunle", "George", "Thomas", "Wesley")
b_age = (19, 16, 18, 17, 20, 19, 16, 18, 17, 19)
b_height = (5.7, 5.9, 5.8, 6.1, 5.9, 5.5, 6.1, 5.4, 5.8, 5.7)
b_scores = (74, 87, 75, 68, 66, 78, 87, 98, 54, 60)
b_gender = ["male"] * 10

# combine all data
all_names = g_names + b_names
all_ages = g_age + b_age
all_heights = g_height + b_height
all_scores = g_scores + b_scores
all_genders = g_gender + b_gender

# create a dataframe
df = pd.DataFrame({
    "Name": all_names,
    "Age": all_ages,
    "Height": all_heights,
    "Scores": all_scores,
    "Gender": all_genders
})

#Display the table
print("Complete Student Data")
print(df)

```

Algorithm:
Create separate lists for girls' and boys' data

Combine all data into a single dataset with a "Gender" column

Create a pandas DataFrame to display the data in tabular form

Print the table showing Name, Age, Height, Score, and Gender
