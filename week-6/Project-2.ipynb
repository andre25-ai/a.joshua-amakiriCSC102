```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox

class DeliveryCalculator:
    def __init__(self, root):
        self.root = root
        self.root.title("Delivery Fee Calculator")
        self.root.geometry("350x300")
        
        # Location UI
        tk.Label(root, text="Select Location:", font=("Arial", 11)).pack(pady=5)
        self.location_box = ttk.Combobox(root, values=["PAU", "Epe"], state="readonly", font=("Arial", 11))
        self.location_box.pack(pady=5)
        
        # Weight UI
        tk.Label(root, text="Enter Weight (kg):", font=("Arial", 11)).pack(pady=5)
        self.weight_entry = tk.Entry(root, font=("Arial", 11))
        self.weight_entry.pack(pady=5)
        
        # Calculate Button
        self.calc_btn = tk.Button(root, text="Calculate Fare", command=self.calculate_fare, bg="blue", fg="white", font=("Arial", 11))
        self.calc_btn.pack(pady=20)

    def calculate_fare(self):
        location = self.location_box.get()
        weight_str = self.weight_entry.get()
        
        # Validation
        if not location:
            messagebox.showwarning("Error", "Please select a location.")
            return
        try:
            weight = float(weight_str)
        except ValueError:
            messagebox.showwarning("Error", "Please enter a valid numerical weight.")
            return

        # Core Calculation Logic
        fare = 0
        if location == "PAU":
            fare = 2000 if weight >= 10 else 1500
        elif location == "Epe":
            fare = 5000 if weight >= 10 else 4000
            
        # Display Result
        messagebox.showinfo("Total Delivery Charge", f"The delivery cost to {location} for a {weight}kg package is: ₦{fare:,}")

if __name__ == "__main__":
    root = tk.Tk()
    app = DeliveryCalculator(root)
    root.mainloop()
```
