```python
from abc import ABC, abstractmethod

# Abstract Base Class
class coup_de_ecriva(ABC):
    @abstractmethod
    def Fan_Page(self):
        pass

# Subclasses using Single Inheritance
class FC_Cirok(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to FC_Cirok")

class Madiba_FC(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to Madiba_FC")

class Blue_Jay_FC(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to Blue_Jay_FC")

class TSG_Walker(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to TSG_Walker")

# Main program execution
def main():
    print("--- Coup de Ecriva Club Fan Pages ---")
    club_input = input("Enter the name of the club you support: ").strip()
    
    # Dictionary mapping user input string to the respective Class references
    clubs = {
        "FC_Cirok": FC_Cirok,
        "Madiba_FC": Madiba_FC,
        "Blue_Jay_FC": Blue_Jay_FC,
        "TSG_Walker": TSG_Walker
    }
    
    # Check if the entered club exists in our record
    if club_input in clubs:
        # Instantiate the matching class object and invoke the abstract-implemented method
        club_instance = clubs[club_input]()
        club_instance.Fan_Page()
    else:
        print("Club not found! Please check your spelling (e.g., FC_Cirok, Madiba_FC).")

if __name__ == "__main__":
    main()
```
